```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
timeStamps_raw = pd.read_json('Data Jumping/JumpData_timeStamped.txt', lines=True)
timeDur_raw = pd.read_csv('Data Jumping/TimeData_timeStamped.txt')
```

```python
eeg_raw = pd.read_csv('Data Jumping/sampleEEG.csv')
data = eeg_raw.values
```

```python
eeg_raw.head()
```

```python
times_trialStart = eeg_raw.dropna()
times_trialStart = times_trialStart[times_trialStart['Event Id']==33025.0][0:3]
numRuns = len(times_trialStart.values)
```

```python
times_trialStart.head()
```

```python
#Session2
times_pu = timeStamps_raw.values[1::,0]
times_saw = timeStamps_raw.values[1::,1]
```

```python
timeStamps_raw.head()
```

```python
times_saw_all = []
for i in range(numRuns):
    times_saw_all.append(times_saw[i][1::2] + times_trialStart.values[i,0])
times_saw_all = np.asarray(times_saw_all).ravel()
```

```python
times_pu_all = []
for i in range(numRuns):
    times_pu_all.append(times_pu[i][1::2] + times_trialStart.values[i,0])
times_pu_all = np.asarray(times_pu_all).ravel()
```

```python
print(times_pu_all)
print(times_saw_all)
```

```python
samplingRate = 64
times_saw_s = np.add((np.multiply(samplingRate,times_saw_all)),samplingRate)
times_pu_s = np.add((np.multiply(samplingRate,times_pu_all)),samplingRate)
```

```python
times_trialStart.index
```

```python
secs = 70

fig = plt.figure(figsize=(20,10))
plt.plot(data[0:secs*samplingRate,2], label='raw EEG (repr. selection: F3)')
plt.plot(data[0:secs*samplingRate,3]+10, label='raw EEG (repr. selection: F4)')
plt.vlines(times_trialStart.index,-5,15, color='red', label='Trial start')
plt.vlines(times_saw_s[:],-5,15, color='green', label='Saws')
plt.vlines(times_pu_s[:], -5,15, color='orange', label='Powerups')
plt.legend()
plt.xticks(np.linspace(0,secs*samplingRate,11), np.linspace(0,secs,11))
plt.yticks([0,10],['F3','F4'])
plt.xlabel('Trial time /sec')
plt.ylabel('EEG raw data /mV')
plt.ylim([-5,15])
#plt.title('Recordings and annotations of three consecutive runs')
plt.show()
```
